```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')

analisis = pd.read_csv("avocado.csv")

analisis.head()
```

```python
analisis.describe()
```

```python
print(analisis.groupby('year').size())
```

```python
analisis.drop(['year'],1).hist(color = 'plum')
plt.show()
```

```python
sb.pairplot(analisis.dropna(), hue='year',height=4,
            vars=["Total Bags","Small Bags","Large Bags", "XLarge Bags"],
            kind='scatter')
```

```python
X = np.array(analisis[["Total Bags","Small Bags","Large Bags", "XLarge Bags"]])
y = np.array(analisis['year'])
X.shape
```

```python
fig = plt.figure()
ax = Axes3D(fig)
colores=['blue','red','green','blue','cyan','yellow','orange','black','pink','brown','purple']
asignar=[]
for row in y:
    asignar.append(colores[row-2015])
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
```

```python
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
score
plt.plot(Nc,score,color='indigo')
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
# Sacar centroids aleatoriamente
# k = 3
kmeans = KMeans(n_clusters = 3).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
# Predecir los clusters
labels = kmeans.predict(X)
# Los cluser centers
C = kmeans.cluster_centers_
colores=['plum','pink','orchid']
asignar=[]
for row in labels:
    asignar.append(colores[row])
 
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c = asignar,s = 60)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c = colores, s = 1000)
```

```python
f1 = analisis['Total Bags'].values
f2 = analisis['Small Bags'].values
 
plt.scatter(f1, f2, c=asignar, s=70)
plt.scatter(C[:, 0], C[:, 1], marker='*', c=colores, s=1000)
plt.show()
```

```python
# Getting the values and plotting it
f1 = analisis['Total Bags'].values
f2 = analisis['Large Bags'].values
 
plt.scatter(f1, f2, c=asignar, s=70)
plt.scatter(C[:, 0], C[:, 2], marker='*', c=colores, s=1000)
plt.show()
```

```python
f1 = analisis['Total Bags'].values
f2 = analisis['XLarge Bags'].values
 
plt.scatter(f1, f2, c=asignar, s=70)
plt.scatter(C[:, 1], C[:, 2], marker='*', c=colores, s=1000)
plt.show()
```
